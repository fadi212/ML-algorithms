```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv(r'E:\Machine Learning A-Z Template Folder\Churnn_Modelling.csv')
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```python
X.shape, y.shape
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lab_x= LabelEncoder()
X[:, 1] = lab_x.fit_transform(X[:, 1])
```

```python
lab_x2= LabelEncoder()
X[:, 2] = lab_x.fit_transform(X[:, 2])
```

```python
one_hot = OneHotEncoder(categorical_features=[1])
X = one_hot.fit_transform(X).toarray()
```

```python
#avoiding dumy variables trap
X = X[:,1:]
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Making ANN
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#initialising ANN
classifier = Sequential()
```

```python
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation='relu', input_dim=11))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation='relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
#softmax for more than two output categories
```

```python
#Compiling the ANN
# loss function is same as in Logistic regression
classifier.compile(optimizer = 'adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size= 10, epochs=100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Evaluating the ANN

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation='relu', input_dim=11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation='relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss= 'binary_crossentropy', metrics=['accuracy'])
    return classifier  
```

```python
classifier = KerasClassifier(build_fn=build_classifier, batch_size=10, nb_epoch=100)
accuracies = cross_val_score(estimator=classifier, X = X_train, y=y_train, cv=10, n_jobs=-1)
```

```python
accuracies
```

```python
mean = accuracies.mean()
variance = accuracies.std()
```

# Improving the ANN with regularization
You can study about Dropout regularization [here](https://machinelearningmastery.com/dropout-regularization-deep-learning-models-keras/) and how it helps the model in avoiding overfitting

```python
from keras.layers import Dropout
```

```python
classifier = Sequential()
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation='relu', input_dim=11))
classifier.add(Dropout(rate=0.1))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation='relu'))
classifier.add(Dropout(rate=0.1))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss= 'binary_crossentropy', metrics=['accuracy'])
#softmax for more than two output categories
```

```python
classifier.fit(X_train,y_train, batch_size=10, epochs=100)
```

**Hyper Parameters Tuning**

```python
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation='relu', input_dim=11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation='relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss= 'binary_crossentropy', metrics=['accuracy'])
    return classifier
```

```python
classifier = KerasClassifier(build_fn=build_classifier)
parameters = {'batch_size':[25, 32],
             'epochs':[100,500],
             'optimizer':['adam','rmsprop']}
grid_search = GridSearchCV(estimator=classifier,param_grid=parameters, scoring='accuracy', cv=10, n_jobs=-1)
```

```python
grid_search = grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

So we have got the final model, which has the following parameters as batch_size of 25, epochs = 500 and optimizer as 'rmsprop'.
